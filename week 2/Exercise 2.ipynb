```python
def euler_method(f, x0, t0, tf, dt):
    
    time_values = [t0]
    x_values = [x0]

    t = t0
    x = x0

    while t < tf:    
        x = x + dt * f(t, x)
        t += dt

        time_values.append(t)
        x_values.append(x)

    return time_values, x_values
```

```python
import matplotlib.pyplot as plt

def f(t, x):
    return x**2   

t0 = 0
tf = 1
x0 = 1
dt = 0.1

time_values, x_values = euler_method(f, x0, t0, tf, dt)

# Plot the results
plt.plot(time_values, x_values, label='x(t)')
plt.xlabel('Time (t)')
plt.ylabel('Value of x')
plt.title('Euler Method Solution')
plt.legend()
plt.grid(True)
plt.show()
```
