(1).Implement a function leapfrogadvect that takes a grid vector x, a time step size dt, a final time T, and an initial condition rho_0 on the grid x, and returns the final grid function rho, similar to the implemented function for the FTCS scheme.


```python
import numpy as np

def leapfrogadvect(x, dt, T, rho0):
    # Compute spatial step size
    dx = x[1] - x[0]

    # Initial condition; set up future time array
    u0 = rho0.copy()
    u1 = u0.copy()
    u2 = u0.copy()

    # Compute nu for the first time step (t = 0)
    nu = dt / (2 * dx)
    
    # Time loop
    for t in np.arange(0, T + dt, dt):
        if t == 0:
            # Compute new values using updated nu for t = 0
            u1[1:len(x)-1] = u0[1:len(x)-1] - nu * (u0[2:len(x)] - u0[0:len(x)-2])
            u1[0] = u0[0] - nu* (u0[1] - u0[len(x)-1])
            u1[len(x)-1]= u0[len(x)-1] - nu* (u0[0]-u0[len(x)-2])
            u0=u1.copy()
            
        else:
            # Update nu for the rest of the time after t = 0
            nu = dt / dx
            # Compute new values using updated nu for t > 0
            u2[1:(len(x)-1)]= u0[1:(len(x)-1)] - nu* (u1[2:len(x)]-u1[0:(len(x)-2)])
            u2[0] = u0[0] - nu* (u1[1] - u1[len(x)-1])
            u2[len(x)-1]= u0[(len(x)-1)] - nu* (u1[0]-u1[len(x)-2])

            # Update u0 for the next iteration, using u2
            u0 = u1.copy()
            u1 = u2.copy()

    # Set output value to the last updated array
    rho = u0
    return rho

from numpy import pi, exp, cos
def init1(x):
    u0=exp(-(20*x)**2/2)
    return u0 
#rho = leapfrogadvect(x, dt, T, rho0)
#print(rho)
```


b) Note that your scheme is still unstable, similar to the FTCS scheme. However, it can be
stabilized by implementing a so-called Asselin Filter. To realize it, the update expression
ρj−1 ← ρj is replaced by: ••
ρj−1 ←ρj +a ρj+1 −2ρj +ρj−1. •••••
Here 0 ≤ a < 1 is a filter parameter. Note that ρj+1 should be known at this stage of the •
algorithm, i.e., this update takes place at the start of each time step, when old values are overwritten by those of the previous iteration. If a = 0, we obtain the original (unstable) leap-frog scheme. For a ≈ 1 heavy damping can be expected.
Implement this simple stabilization for the leap-frog scheme and test it with e.g. a = 0.05.


```python
def leapfrogadvect(x, dt, T, rho0, a):
    # Compute spatial step size
    dx = x[1] - x[0]

    # Initial conditions
    u0 = rho0.copy()
    u1 = u0.copy()
    u2 = u0.copy()

    # Compute nu for the first time step (t = 0)
    nu = dt / (2 * dx)
    
    # Time loop
    for t in np.arange(0, T + dt, dt):
        if t == 0:
            # Compute new values using updated nu for t = 0
            u1[1:len(x)-1] = u0[1:len(x)-1] - nu * (u0[2:len(x)] - u0[0:len(x)-2])
            u1[0] = u0[0] - nu* (u0[1] - u0[len(x)-1])
            u1[len(x)-1]= u0[len(x)-1] - nu* (u0[0]-u0[len(x)-2])
            u0=u1.copy()   
            
        else:
            # Update nu for the rest of the time after t = 0
            nu = dt / dx
            # Compute new values using updated nu for t > 0
            u2[1:(len(x)-1)]= u0[1:(len(x)-1)] - nu* (u1[2:len(x)]-u1[0:(len(x)-2)])
            u2[0] = u0[0] - nu* (u1[1] - u1[len(x)-1])
            u2[len(x)-1]= u0[(len(x)-1)] - nu* (u1[0]-u1[len(x)-2])
            
            # Asselin filter application to stabilize the scheme
            u0 = u1 + a*(u2 - (2 * u1) + u0)
            u1=u2.copy()
            

    # Set output value to the last updated array
    rho = u2
    return rho
```

```python
def testlf():
    import numpy as np
    from numpy.linalg import norm
    import matplotlib.pyplot as plt

    # Initial values/settings
    dx = 1. / 100.              # spatial step size
    courant = 0.5               # Courant number
    dt = courant * dx           # time step size
    Tend = 0.5                  # final time
    x = np.arange(-1, 1 + dx, dx)  # spatial grid
    a = 0.05                   # Asselin filter coefficient
    
    # Define initial condition function
    def init1(x):
        return np.exp(-(20 * x) ** 2 / 2)

    # Initial conditions
    u0 = init1(x)
    
    # Leapfrog advection scheme with Asselin filter
    uleapfrog = leapfrogadvect(x, dt, Tend, u0, a)
    
    # Compute exact solution to compare with (assuming periodic boundary conditions)
    uexact = init1(x-Tend)
    
    # Plot the result, the initial condition, and the exact solution
    fig, ax = plt.subplots()
    ax.plot(x, u0, linewidth=2, color=[0.7, 0.7, 0.7], label='Initial condition')
    ax.plot(x, uexact, linewidth=2, color=[0.5, 0.7, 0.4], label='Exact solution')
    ax.plot(x, uleapfrog, linewidth=2, color='red', label='Leapfrog solution with Asselin Filter')
    ax.legend(loc='upper left')
    ax.set_title('Linear Advection, DX=' + '%6.4f' % dx + ', DT=' + '%6.4f' % dt)
    ax.set_xlabel('x')
    ax.set_ylabel('u')
    
    # Compute error norms:
    inftcs = norm(uexact - uleapfrog, np.inf) / norm(uexact, np.inf)
    print('*** INFO: relative error in inf-norm ***')
    print('          Leapfrog method: ' + '%6.4f' % inftcs)
    twotcs = norm(uexact - uleapfrog) / norm(uexact)
    print('*** INFO: relative error in two-norm ***')
    print('          Leapfrog method: ' + '%6.4f' % twotcs)
    
    plt.show()

if __name__ == "__main__":
    testlf()
```
